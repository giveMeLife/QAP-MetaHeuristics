```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
instance = "kra32.dat"
total_times = []
files = []
dfs = []
fig, ax = plt.subplots(nrows = 2, figsize=(10, 10), sharex=True)
for f in os.listdir("logs/monitor/"):
    if f[-3:] == "txt":
        if instance not in f:
            continue
        files.append(f)
        part = f.split("_")[2]
        df = pd.read_csv("logs/monitor/%s" % f, sep="\s+")
        df.drop(["(%)", "Real", "(MB)", "Virtual", "(MB).1"], axis=1, inplace=True)
        df.rename({'#': "time", "Elapsed": "cpu", "time": "mem_real", "CPU": "mem_virtual"}, axis=1, inplace=True)
        if "SA" in f:
            c = 'coral'
        else:
            c = 'skyblue'
        idxmax = df.time.idxmax()
        ax[0].plot(df.time, df.mem_real, c=c, alpha=0.2)
        ax[0].scatter(df.time.loc[idxmax], df.mem_real.loc[idxmax], c=c, marker="o")
        ax[1].plot(df.time, df.cpu, c=c, alpha=0.2)
        ax[1].scatter(df.time.loc[idxmax], df.cpu.loc[idxmax], c=c,  marker="o")
        ax[1].set_xlabel("Tiempo [s]")
        ax[0].set_ylabel("Memoria [MB]")
        ax[1].set_ylabel("CPU (%)")
        dfs.append(df)
        #print(f"{f} {df.time.max()/60}s")
        total_times.append(df.time.max())
print(f"Max: {max(total_times)/60} mins")
print(f"Min: {min(total_times)/60} mins ")
print(f"Total files {len(files)}")
plt.savefig(f"resources_{instance}.png")
```
